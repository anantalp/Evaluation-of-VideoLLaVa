```python
def similarity(s1, s2):
    from sentence_transformers import SentenceTransformer, util
    model = SentenceTransformer('distilbert-base-nli-mean-tokens')
    s=[]
    s.append(s1)
    s.append(s2)
    sentence_embeddings = model.encode(s)
    # print(sentence_embeddings)

    sim=util.pytorch_cos_sim(sentence_embeddings[0], sentence_embeddings[1])
    # print(sim)
    return sim

similarity('hey', 'hi')
```

```python
import json

with open('msvd.jsonl', 'r') as json_file:
    json_list = list(json_file)
    numofrecords=len(json_list)
    print(numofrecords)
i=0
currentpred=0
for json_str in json_list:
    result = json.loads(json_str)
    answer=result['answer']
    pred=result['pred']
    sim=similarity(answer,pred)
    # print(sim)
    if sim>0.5:
        currentpred+=1
    # print(currentpred)
    # print(numofrecords)
    # print(answer)
    # print(pred)
    i+=1
    if i>50:
         break
    # print(isinstance(result, dict))

accuracy=currentpred/numofrecords
print(accuracy)
```
