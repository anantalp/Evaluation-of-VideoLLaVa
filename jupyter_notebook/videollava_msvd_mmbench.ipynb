```python
#mmbench
import pandas as pd

def get_accuracy(file_name):
    df = pd.read_excel(file_name)
    accuracy, correctpred=0,0
    numofrecords=len(df)
    print(numofrecords)
    for index, row in df.iterrows():
        if row['prediction'] == row['answer']:
            correctpred+=1
    accuracy = correctpred/numofrecords
    return accuracy

print("mmbench accuracy")
get_accuracy('mmbench.xlsx')
```

```python
def similarity(s1, s2):
    from sentence_transformers import SentenceTransformer, util
    model = SentenceTransformer('distilbert-base-nli-mean-tokens')
    s=[]
    s.append(s1)
    s.append(s2)
    sentence_embeddings = model.encode(s)
    # print(sentence_embeddings)

    sim=util.pytorch_cos_sim(sentence_embeddings[0], sentence_embeddings[1])
    # print(sim)
    return sim

similarity('hey', 'hi')
```
