```python
def similarity(s1, s2):
    from sentence_transformers import SentenceTransformer, util
    model = SentenceTransformer('distilbert-base-nli-mean-tokens')
    s=[]
    s.append(s1)
    s.append(s2)
    sentence_embeddings = model.encode(s)
    # print(sentence_embeddings)

    sim=util.pytorch_cos_sim(sentence_embeddings[0], sentence_embeddings[1])
    # print(sim)
    return sim

similarity('hey', 'hi')
```

```python
import json
with open('answers_gpt4.jsonl', 'r') as json_file:
    gpt4 = list(json_file)
    gpt_num=len(gpt4)
    print(gpt_num)

with open('llavabench.jsonl', 'r') as json_file:
    json_list = list(json_file)
    numofrecords=len(json_list)
    print(numofrecords)
# i=0
currentpred=0
for json_str in json_list:
    result = json.loads(json_str)
    # print(result['question_id'])
    # print(result['text'])
    for j in gpt4:
        result2 = json.loads(j)
        if result['question_id']==result2['question_id']:
            gt=result2['text']
            pred=result['text']
            # print(gt)
            # print(pred)
            sim=similarity(gt, pred)
            print(sim)
            if sim>0.75:
                currentpred+=1
                break
            # print(result['question_id'])
            # print(result['text'])
            # print(result2['text'])
print(currentpred)
accuracy=currentpred/numofrecords
print(accuracy)
```
